```python
import numpy as np
import random
import math

class Node(object):
    def __init__(self, data):
        self.data = data
        self.left_child = None
        self.right_child = None

    def add_left_child(self, obj):
        self.left_child = obj

    def add_right_child(self, obj):
        self.right_child = obj
def calc_default(train,bestIG):
    count0=0
    count1=0
    for i in range(len(train)):
        if train[i][bestIG]==0:
            count0+=1
        else:
            count1+=1
    return count0,count1  
#splitting the values as per the information gain index
def destree(train,last,Initialentropy,spam,notspam):
    bestIG=calculate_entropy(train,last,Initialentropy)
    tree = Node(bestIG)
    count0,count1=calc_default(train,bestIG)
    if count0==0:
        return Node(0)
    elif count1==1:
        return Node(1)
    
    value1=[]
    value0=[]
    for i in range(len(train)):
        if train[i][bestIG]==1:
            value1.append(train[i])
        elif train[i][bestIG]==0:
            value0.append(train[i])
    value1 = np.array(value1) 
    last1 = value1[:,value1.shape[1] - 1:]
    value0 = np.array(value0)
    last0 = value0[:,value0.shape[1] - 1:]
    print(last1)
        
                   
#grouping the values to 0 and 1                
def groupify(train,test):
    
    for i in range(len(train)):
        for j in range(train.shape[1]):
            if train[i][j]>0:
                train[i][j]=1
            else:
                train[i][j]=0
    for i in range(len(test)):
        for j in range(test.shape[1]):
            if test[i][j]>0:
                test[i][j]=1
            else:
                test[i][j]=0
                
    return(train,test)           
                
def standardise(data):
    mean= np.mean(data, axis=0)
    std = np.std(data, axis=0, ddof=1)
    Stdfinal = (data-mean)/std
    return Stdfinal

def checkZero(data):
    if data==0:
        data=.0001
    return data    
#calculating entropy    
    
def calculate_entropy(train,last,Initialentropy):
    spam,notspam=makeGroup(train,last)
    p0=[]
    p1=[]
    sumEntropy=[]
    informationGain=[]
    lens0=[0]*spam.shape[1]
    lens1=[0]*spam.shape[1]
    lenn0=[0]*spam.shape[1]
    lenn1=[0]*spam.shape[1]
    print(spam.shape)
    for i in range(len(spam)):
        for j in range(spam.shape[1]):
            if spam[i][j]==0:
                lens0[j]+=1
            elif spam[i][j]==1:
                 lens1[j]+=1
        
    for i in range(len(notspam)):
        
        for j in range(notspam.shape[1]):
            if notspam[i][j]==0:
                lenn0[j]+=1
            elif notspam[i][j]==1:
                 lenn1[j]+=1
    for i in range(spam.shape[1]):
        p0.append(0)
        p1.append(0)
        lenn0[i]=checkZero(lenn0[i])
        lenn1[i]=checkZero(lenn1[i])
        lens0[i]=checkZero(lens0[i])
        lens1[i]=checkZero(lens1[i])
        
        p0[i]=-(lens0[i]/(lens0[i]+lenn0[i]))*math.log(lens0[i]/(lens0[i]+lenn0[i]))-(lenn0[i]/(lens0[i]+lenn0[i]))*math.log(lenn0[i]/(lens0[i]+lenn0[i]))
        p1[i]=-(lens1[i]/(lens1[i]+lenn1[i]))*math.log(lens1[i]/(lens1[i]+lenn1[i]))-(lenn1[i]/(lens1[i]+lenn1[i]))*math.log(lenn1[i]/(lens1[i]+lenn1[i]))
#      
    for i in range(spam.shape[1]):
        sumEntropy.append(((lens0[i]+lenn0[i])/(len(spam)+len(notspam))*p0[i])+((lens1[i]+lenn1[i])/(len(spam)+len(notspam))*p1[i]))
        informationGain.append(Initialentropy-sumEntropy[i])
    maxValue=max(informationGain)
    print("Index of best information gain",informationGain.index(maxValue))
    return informationGain.index(maxValue)
    
#divide to spam notspam

def makeGroup(train,last):
    spam=[]
    notspam=[]
    for i in range(len(train)):
        if last[i]==1:
            spam.append(train[i])
        elif last[i]==0:
            notspam.append(train[i])
    spam = np.array(spam)  
    notspam = np.array(notspam)
    return spam,notspam
# def countgroup():
#     return len(group)

def divideData(data):
    train_pct_index = int((.68) * len(data))
    train,test = data[:train_pct_index], data[train_pct_index:]
    last = train[:, train.shape[1] - 1:]
    train = train[:, :train.shape[1] - 1]
    ActualResult = test[:, test.shape[1] - 1:]
    test = test[:, :test.shape[1] - 1]
    return train,test,last,ActualResult

def main():
    data = np.genfromtxt('./untitled2.txt',delimiter=',')
    np.random.seed(0)
    np.random.shuffle(data)
    train,test,last,ActualResult=divideData(data)
    train=standardise(train)
    test=standardise(test)
    
    train,test=groupify(train,test)
    spam,notspam=makeGroup(train,last)
    ProbSpam=len(spam)/(len(spam)+len(notspam))
    ProbNotSpam=len(notspam)/(len(spam)+len(notspam))
    Initialentropy=-ProbSpam*math.log(ProbSpam,2)-ProbSpam*math.log(ProbSpam,2)
    destree(train,last,Initialentropy,spam,notspam)
    
    
    
if __name__ == "__main__":
    main()
```
